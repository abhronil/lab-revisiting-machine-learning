```python
drop_list=[]
```

```python
drop_list.append("OSOURCE")
```

```python
drop_list.append("ZIP")
```

```python
null = pd.DataFrame(df.isna().sum()*100/len(df), columns=['percentage'])
```

```python
null
```

```python
null.sort_values('percentage', ascending = False).head(50)
```

```python
for index, row in null.iterrows():
    if row['percentage'] > 15:
        drop_list.append(index)
```

```python
drop_list
```

```python
for col in drop_list:
    df = df.drop(col, axis=1)
```

```python
print(df['GENDER'].value_counts())
df['GENDER'] = df['GENDER'].fillna('F')
```

```python
print(sorted(df['GENDER'].unique()))
```

```python
def map_gender(value):
    if value not in ['m', 'f']:
        return 'O'
    return value
```

```python
df["GENDER"] = df["GENDER"].applymap(map_gender)
```
